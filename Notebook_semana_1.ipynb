```python
# ENTRADAS: Ancho y Alto en metros
Ancho = input("Ingrese Ancho de la ventana en metros")
Ancho = float(Ancho)
Alto = input("Ingrese el Alto de la ventana en metros")
Alto = float(Alto)

# PROCESO: Calcula lm en pies y av en metros cuadrados
lm = 2 * (Ancho + Alto) * 3.25
av = (Ancho * Alto) * 2

# SALIDA: sale longitud de la madera (lm) en pies y area del vidrio en metros cuadrados
print("longitud de la madera =",lm,"pies de longitud")
print("area de la ventana =",av,"Metros cuadrados")
```
